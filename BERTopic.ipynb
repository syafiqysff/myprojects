```python
import pandas as pd 
import numpy as np
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer

vectorizer_model = CountVectorizer(stop_words="english")
topic_model = BERTopic(vectorizer_model=vectorizer_model)
```

```python
#load data

import pandas as pd 
 
df = pd.read_csv("japan_translated_eng.csv")
df2 = pd.read_csv("florida_translated_eng.csv")
df3 = pd.read_csv("singapore_translated_eng.csv")
```

## Getting neutral and negative reviews. Slice Dataframe with ratings 3 and below

```python
df_neutral_bad = df.loc[df['rating'] <= 3.0]
```

```python
df_neutral_bad.head()
```

```python
df2_neutral_bad = df2.loc[df2['rating'] <= 3.0]
```

```python
df2_neutral_bad.head()
```

```python
df3_neutral_bad = df3.loc[df3['rating'] <= 3.0]
```

```python
df3_neutral_bad.head()
```

## Topic modelling on neutral and negative reviews from Universal Studios Japan

```python
vectorizer_model = CountVectorizer(stop_words="english")
topic_model = BERTopic(vectorizer_model=vectorizer_model, verbose=True)
 
#convert to list 
docs = df_neutral_bad.translated.to_list()
 
topics, probabilities = topic_model.fit_transform(docs)
```

```python
# top topics
topic_model.get_topic_freq()
```

```python
# selecting 1 topic
topic_model.get_topic(1)
```

```python
# visualize topics
topic_model.visualize_topics()
```

```python
# visualize terms
topic_model.visualize_barchart(n_words=10, width=300, height=540)
```

```python
# visualize topic similarity
topic_model.visualize_heatmap()
```
